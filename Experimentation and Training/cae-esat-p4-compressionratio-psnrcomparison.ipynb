```python
import matplotlib.pyplot as plt
import os
import pickle

os.listdir('../input/cae-esat-p4-compressionrate')
foldernames = []
for i in os.listdir('../input/cae-esat-p4-compressionrate'):
    if i[0:5]=='data_':
        foldernames.append(i)

plt_1 = plt.figure(figsize=(10, 7))
plt.rcParams.update({'font.size': 16})

for i in foldernames:
    filehandler = open('../input/cae-esat-p4-compressionrate/'+i+'/logger.obj', 'rb') 
    logger = pickle.load(filehandler)
    plt.plot(logger.epochwisepsnrval, label = i.split('_')[1])

plt.legend()
plt.xlabel('No. of Epochs')
plt.ylabel('PSNR (in dB)')
plt.savefig('CompressionRate_comp_psnr.png')
plt.show()
```
